```python
import pandas as pd 
import warnings 
warnings.filterwarnings("ignore")
names_score = ['id','forum_id','score_rank']
# 导入学习成绩数据
score_train = pd.read_csv("./train/score_train.txt",header=None,names = names_score,index_col = None)
score_test = pd.read_csv("./test/score_final_test.txt",header=None,names = names_score,index_col = None)
score_data = pd.concat([score_train,score_test])
```

```python
%matplotlib inline
import pandas as pd
import numpy as np

#Set some Pandas options
pd.set_option('max_columns', 30)
pd.set_option('max_rows', 20)
```

```python
libray_names = ['id','door_num','lib_time']
lib_test = pd.read_csv('./test/library_final_test.txt',header=None,names=libray_names)
lib_train = pd.read_csv('./train/library_train.txt',header=None,names=libray_names)
lib_data = pd.concat([lib_test,lib_train], ignore_index=False)
del lib_test,lib_train,libray_names
lib_data
```

#将数据时间按月进行解析

```python
lib_data['lib_month']=pd.to_datetime(lib_data['lib_time']).apply(lambda x: x.strftime('%Y-%m'))
# borrow_data.drop('borrow_time',1, inplace=True)
lib_data.head()
```

```python
lib_use = lib_data.groupby(['id','lib_month']).count().unstack('lib_month')['lib_time']
```

```python
lib_use.head()
```

```python
lib_use.loc[0].plot(kind='bar')
```

```python
lib_times = lib_data.groupby(['id'])['lib_time'].count()
```

```python
df = pd.DataFrame({'id':lib_times.index,'lib_times':lib_times.values})
```

```python
df.head()
```

```python
df.agg(['max','min','sum','mean'])
```

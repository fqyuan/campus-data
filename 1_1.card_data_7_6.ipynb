```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
#Set some Pandas options
pd.set_option('max_columns', 30)
pd.set_option('max_rows', 20)
```

```python
#Set colunm index and index value
column_index = ('id','con_cate','con_addr','con_content','con_time','con_val','rem_count')
card_data_test = pd.read_csv('./test/card_final_test.txt', header=None)
card_data_train = pd.read_csv('./train/card_train.txt', header=None)

card_data_test.columns = column_index
card_data_train.columns = column_index

# card_data_test.set_index('id', inplace=True)
# card_data_train.set_index('id', inplace=True)

#Combine two dataframes in pandas
# bigdata = data1.append(data2, ignore_index=True)
card_data = pd.concat([card_data_train, card_data_test], ignore_index=False)

#Release memory
del card_data_test,card_data_train, column_index
# del card_data_train
```

```python
#Drop duplicates
# card_clean = card_data.sort_values('con_val',ascending=True).drop_duplicates(subset=['id','con_time']).sort_index()
card_clean = card_data.sort_values('id',ascending=True).drop_duplicates(subset=['id','con_time'])
#Delete variable to release memory
del card_data
# card_clean.head()
```

#首先将数据的时间解析为Y/M格式

```python
card_clean['month_time']=pd.to_datetime(card_clean['con_time']).apply(lambda x: x.strftime('%Y%m'))
card_clean.drop('con_time',1, inplace=True)
card_clean.head()
```

```python
# card_clean.sort_values(['id','con_val'],ascending=True, inplace=True)
card_clean.head()
```

```python
#Deal with con_way column
card_clean['con_content'] = card_clean['con_content'].astype('category')
card_clean['con_content'].describe()
```

```python
print(card_clean['con_content'].unique())
```

```python
card_clean.isnull().sum()
```

```python
#Set negative value to 0 and remove that row
card_clean = card_clean.clip(lower=0)
# type(card_clean)
card_clean = card_clean[~(card_clean==0).any(axis=1)]
#Replace NaN value with 'Default'
card_clean['con_content'].fillna('食堂', inplace=True)
```

#这里得到的是关于总的学生消费金额数据

```python
#Total cost of each student, and mean cost of each student
card_clean_sum_by_id = card_clean.groupby('id')['con_val'].sum()
card_clean_mean_by_id = card_clean.groupby('id')['con_val'].sum().mean()
total_cost = card_clean.groupby('id')['con_val'].sum().sum()
card_clean_sum_by_id.head()
```

```python
card_clean_mean_by_id
```

```python
total_cost
```

```python
#Cost of each student by id & category
card_clean_sum_by_ID_type = card_clean.groupby(['id','con_content'])['con_val'].sum().unstack('con_content')
card_clean_sum_by_ID_type
```

```python
card_clean_sum_by_ID_type.fillna(0, inplace=True)
card_clean_sum_by_ID_type.head()
```

```python
#Combine data together
card_id_not_by_month = pd.concat([card_clean_sum_by_id, card_clean_sum_by_ID_type],axis=1)
card_id_not_by_month.head()
```

#这里得到的是按月份按类别的消费数据

```python
card_clean_sum_by_ID_type_time = card_clean.groupby(['id','month_time','con_content'])['con_val'].sum().unstack('con_content')
```

```python
card_clean_sum_by_ID_type_time.fillna(0, inplace=True)
card_clean_sum_by_ID_type_time.head()
```

```python
card_clean_sum_by_ID_type_time_sum = card_clean.groupby(['id','month_time'])['con_val'].sum()
```

```python
card_clean_sum_by_ID_type_time_sum.head()
```

```python
#Combine data together
card_id_by_month = pd.concat([card_clean_sum_by_ID_type_time_sum,card_clean_sum_by_ID_type_time],axis=1)
card_id_by_month.head(10)
```
